# 5. Exploratory Data Analysis:

```python
# Encoding categorical data.
data = pd.get_dummies(data,drop_first=True)
```

```python
# Let's see how data is distributed for every column.
plt.figure(figsize = (20, 25), facecolor = 'yellow')
plotnumber = 1
for column in data:
    if plotnumber <= 5 : # As there are 9 columns in the data.
        ax = plt.subplot(3,3,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
        plotnumber+=1
plt.show()
```

```python
# Correlation matrix
sns.set(rc={'figure.figsize':(12.7,9.27)})
sns.heatmap(data.corr(),annot=True,fmt='.2f')
plt.show()
```

```python
plt.figure(figsize=(15,7))
data.corr()["price"].sort_values(ascending=False).drop(["price"]).plot.bar()
plt.xlabel("Feature", fontsize= 14)
plt.ylabel("correlation with Target column", fontsize = 18)
plt.title("Correlation of Fetures with the target column", fontsize=25)
plt.show()
cor=pd.DataFrame(data.corr()["price"].sort_values(ascending=False)).drop(["price"])
print("KEY OBSERVATIONS:\n ")
cor
```
